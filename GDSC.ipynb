```python
df.replace([np.inf, -np.inf], np.nan, inplace=True)
```

```python
df.info()
```

```python
from sklearn.preprocessing import LabelEncoder
l = LabelEncoder()
for i in df.columns:
    if df[i].dtype == 'object':
      df[i] = l.fit_transform(df[i])
```

```python
df.info()
```

```python
df.dropna(inplace=True)
X = df.iloc[:, :86]
Y = df.iloc[:, 86]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier()
dt.fit(X_train, Y_train)

num_features_to_select = 15
feature_importances = dt.feature_importances_
selected_features = X_train.columns[np.argsort(feature_importances)[-num_features_to_select:]]

print("Features Selected")
print()
for i in selected_features:
    print(i)
```

```python
X_train.reset_index(drop=True, inplace=True)
Y_train.reset_index(drop=True, inplace=True)

new_X_train = X_train[selected_features]
new_X_test = X_test[selected_features]
new_XY_train = pd.concat([new_X_train, Y_train], axis = 1)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, balanced_accuracy_score

lr = LogisticRegression()
lr.fit(new_X_train, Y_train)
Y_pred = lr.predict(new_X_test)

accuracy = accuracy_score(Y_test, Y_pred)
precision = precision_score(Y_test, Y_pred)
recall = recall_score(Y_test, Y_pred)
f1 = f1_score(Y_test, Y_pred)
confusion = confusion_matrix(Y_test, Y_pred)
balanced_accuracy = balanced_accuracy_score(Y_test, Y_pred)

print("Accuracy:", accuracy)
print("Precision:", precision)
print("Recall:", recall)
print("F1 Score:", f1)
print("Confusion Matrix:\n", confusion)
print("Balanced Accuracy:", balanced_accuracy)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, balanced_accuracy_score

rf = RandomForestClassifier()
rf.fit(new_X_train, Y_train)
Y_pred = rf.predict(new_X_test)

accuracy = accuracy_score(Y_test, Y_pred)
precision = precision_score(Y_test, Y_pred)
recall = recall_score(Y_test, Y_pred)
f1 = f1_score(Y_test, Y_pred)
confusion = confusion_matrix(Y_test, Y_pred)
balanced_accuracy = balanced_accuracy_score(Y_test, Y_pred)

print("Accuracy:", accuracy)
print("Precision:", precision)
print("Recall:", recall)
print("F1 Score:", f1)
print("Confusion Matrix:\n", confusion)
print("Balanced Accuracy:", balanced_accuracy)
```
